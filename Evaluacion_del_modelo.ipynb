```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.models import load_model
```

```python
data = pd.read_excel('./data_01.xlsx', index_col=0)
data

# Cargar el modelo desde el archivo
model = load_model("modelo_entrenado.h5")
```

```python
input_cols = ['MAF', 'velocidad_motor', 'masa_combustible_inyectado']
output_cols = ['MAP']

X = data[input_cols].values
y = data[output_cols].values

# Normalizar los datos
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)
```

```python
# Obtener los valores reales del sensor MAP
real_MAP = data['MAP'].values
real_MAP = real_MAP.reshape(36068,1)
# Obtener los valores predichos por la red neuronal
predicted_MAP = model.predict(X_scaled)

# Escalar los valores predichos a la misma escala que los reales
predicted_MAP_scaled = scaler_y.inverse_transform(predicted_MAP)

# Calcular el error como un porcentaje
error_percentage = ((predicted_MAP_scaled[:40] - real_MAP[:40]) / real_MAP[:40]) * 100

# Crear la gr√°fica principal
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('MAP (MPa) - Real', color=color)
ax1.plot(real_MAP[:40], color=color, label='MAP (Real)')
ax1.plot(predicted_MAP_scaled[:40], color='tab:blue', label='MAP (Previsto)')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

# Crear el segundo eje y para el error como porcentaje
ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Error (%)', color=color)
ax2.plot(error_percentage[:40], color=color, linestyle='--', label='Error')
ax2.tick_params(axis='y', labelcolor=color)
ax2.legend(loc='upper right')

fig.tight_layout()
plt.show()
```
