```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('./data_01.xlsx', index_col=0)
data
```

```python
data.info()
```

```python
# Filtrar los datos para incluir solo los primeros 60 segundos
data_subset = data[data.index <= 60]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar solo la variable MAP en el eje Y izquierdo
ax.plot(data_subset.index, data_subset['MAP'], color='blue', label='MAP')
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel('MAP (MPa)')

# AÃ±adir leyenda
ax.legend(loc='upper left')

# Mostrar la grÃ¡fica
plt.show()
```

```python
# Filtrar los datos para incluir solo los primeros 60 segundos
data_subset = data[data.index <= 60]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar solo la variable posicion_acelerador en el eje Y derecho
ax.plot(data_subset.index, data_subset['posicion_acelerador'], color='red', label='PosiciÃ³n del Pedal de AceleraciÃ³n')
ax.set_ylabel('PosiciÃ³n del Pedal de Aceler. (%)')
ax.set_xlabel('Tiempo (s)')

# AÃ±adir leyenda
ax.legend(loc='upper left')

# Mostrar la grÃ¡fica
plt.show()
```

```python
# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Graficar MAP en el eje Y izquierdo
ax1.plot(data.index, data['MAP'], color='blue', label='MAP')
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('MAP (MPa)', color='blue')
ax1.tick_params('y', colors='blue')

# Crear un segundo eje Y para posicion_acelerador y masa_combustible_inyectado
ax2 = ax1.twinx()

# Escalar los valores de injected_fuel_mass para ajustar la escala
scaling_factor = 10000  # Ajusta este valor segÃºn tu preferencia
ax2.plot(data.index, data['posicion_acelerador'], color='green', label='Posicion del pedal de aceleraciÃ³n')
ax2.plot(data.index, data['masa_combustible_inyectado'] * scaling_factor, color='red', label='Masa de combustible inyectada (escalado)')
ax2.set_ylabel('Posicion del pedal de aceleraciÃ³n (%)\n Masa de combustible inyectado (g)', color='black')
ax2.tick_params('y', colors='black')

# AÃ±adir leyenda
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

# Mostrar la grÃ¡fica
plt.show()
```
